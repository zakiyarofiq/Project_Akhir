```python
import torch
from PIL import Image
import torchvision.transforms as T
import os
```

```python
model = torch.load("./skin-model_rev.pt", map_location=torch.device('cpu'))
```

```python
# Test on an image
img = Image.open("./gdrive/MyDrive/datatraining/eksim/e3.JPG").convert("RGB")
```

```python
def get_transforms():
    transform = []
    transform.append(T.Resize((512, 512)))
    transform.append(T.ToTensor())
    return T.Compose(transform)
```

```python
tr = get_transforms()
```

```python
img_tensor = tr(img)
```

```python
def predict(model, img, tr, classes):
    img_tensor = tr(img)
    out = model(img_tensor.unsqueeze(0))
    pred, idx = torch.max(out, 1)
    return classes[idx]
```
